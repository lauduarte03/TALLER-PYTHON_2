```python
import pandas as pd
import numpy as np
import wooldridge as wd
```

```python
wage = wd.data('wage1')[['wage','educ','exper']]
wage
```

```python
y = wage ['wage'].to_numpy()
x = wage [['educ','exper']].to_numpy()
u = wage.to_numpy()
```

```python
x
```

```python
y
```

```python
u
```

```python
X = np.c_[np.ones(len(x)),x]
```

```python
X
```

```python
BETAS = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
BETAS #(X^T * X)^-1 * X^T*y #MCO
```

```python
X.dot(BETAS)
```

```python
y_est = X.dot(BETAS)
```

```python
y
```

```python
y - y_est
```

```python
error_est = y - y_est
```

```python
y.shape #n #n-k-1 #k = 2 
```

```python
sum(error_est ** 2) #numerador
```

```python
sum(error_est ** 2) / (y.shape [0]- 2 - 1)
```

```python
sigma_cudr = sum(error_est ** 2) / (y.shape [0]- 2 - 1)
```

```python
np.diag(np.sqrt(sigma_cudr * (np.linalg.inv((X.T.dot(X))))))
```

```python
u_estandar = np.diag(np.sqrt(sigma_cudr * (np.linalg.inv((X.T.dot(X))))))
```

```python
BETAS / u_estandar #t
```

```python
BETAS + 1.96*(u_estandar) #Int.Conf sup
```

```python
BETAS - 1.96*(u_estandar) #Int.Conf inf
```

Para todos los intervalos de confianza son estadisticamente significativos dado que ninguno pasa por cero.
